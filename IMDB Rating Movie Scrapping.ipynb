# Import Library

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# URL

```python
url='https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=5aab685f-35eb-40f3-95f7-c53f09d542c3&pf_rd_r=EJWMJCDS54SX3PZPR7G1&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1'
```

# Requests

```python
page=requests.get(url)
page.status_code
soup=BeautifulSoup(page.content,'html.parser')
```

# Result

```python
movies=soup.find_all('div',class_='lister-item mode-advanced')
len(movies)
```

# Put Evrything together inside a foor loop


```python
title=[]
year=[]
certificate=[]
duration=[]
genre=[]
rating=[]
metascore=[]
director=[]
vote=[]
gross=[]


for movie in movies:
    #title
    try:
        title.append(movie.find('h3',class_='lister-item-header').a.text)
    except:
        title.append('n\a')
        
    #year    
    try:
        year.append(movie.find('span',class_='lister-item-year').text.replace('(','').replace(')',''))
    except:
        year.append('n\a')
        
    #certificate
    
    try:
        certificate.append(movie.find('span',class_='certificate').text)
    except:
        certificate.append('n\a')
        
    #duration
    try:
        duration.append(movie.find('span',class_='runtime').text)
    except:
        duration.append('n\a')
    
    #genre
    
    try:
        genre.append(movie.find("span",class_='genre').text.replace('\n','').replace(' ',''))
    except:
        genre.append('n\a')
    
    #rating
    
    try:
        rating.append(movie.find('div',class_='ratings-bar').strong.text)
    except:
        rating.append('n\a')
        
    #metascore
    
    try:
        metascore.append(movie.find('span',class_='metascore').text.replace(' ',''))
    except:
        metascore.append('n\a')
    
    #director
    
    try:
        director.append(movie.find('p',class_="").a.text)
    except:
        director.append('n\a')
        
    #votes
    
    try:
        value= movie.find_all('span',attrs={"name":'nv'})
        votes=value[0].text
        vote.append(votes)
        gross_sale=value[1].text
        gross.append(gross_sale)
    except:
        vote.append('n\a')
        gross.append('n\a')
        
```

```python
a = {'Movie_name':title,'Year':year,'Certificate':certificate,
                         'Runtime':duration,'Genre':genre,'Rating':rating,'Metascore':metascore,'Director':director,
                         'Votes':vote,'Gross_sale':gross}
df = pd.DataFrame.from_dict(a, orient='index')
df = df.transpose()
```

# Scrape Multiple Page

```python

title=[]
year=[]
certificate=[]
duration=[]
genre=[]
rating=[]
metascore=[]
director=[]
vote=[]
gross=[]

for i in range(1,21):
    #website
    url='https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=5aab685f-35eb-40f3-95f7-c53f09d542c3&pf_rd_r=EJWMJCDS54SX3PZPR7G1&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1'+ str(i)
    #request
    page=requests.get(url)
    
    #soup object
    soup=BeautifulSoup(page.content,'html.parser')
    
    #results
    movies=soup.find_all('div',class_='lister-item mode-advanced')
    

    for movie in movies:
        #title
        try:
            title.append(movie.find('h3',class_='lister-item-header').a.text)
        except:
            title.append('n\a')

        #year    
        try:
            year.append(movie.find('span',class_='lister-item-year').text.replace('(','').replace(')',''))
        except:
            year.append('n\a')

        #certificate

        try:
            certificate.append(movie.find('span',class_='certificate').text)
        except:
            certificate.append('n\a')

        #duration
        try:
            duration.append(movie.find('span',class_='runtime').text)
        except:
            duration.append('n\a')

        #genre

        try:
            genre.append(movie.find("span",class_='genre').text.replace('\n','').replace(' ',''))
        except:
            genre.append('n\a')

        #rating

        try:
            rating.append(movie.find('div',class_='ratings-bar').strong.text)
        except:
            rating.append('n\a')

        #metascore

        try:
            metascore.append(movie.find('span',class_='metascore').text.replace(' ',''))
        except:
            metascore.append('n\a')

        #director

        try:
            director.append(movie.find('p',class_="").a.text)
        except:
            director.append('n\a')

        #votes

        try:
            value= movie.find_all('span',attrs={"name":'nv'})
            votes=value[0].text
            vote.append(votes)
            gross_sale=value[1].text
            gross.append(gross_sale)
        except:
            vote.append('n\a')
            gross.append('n\a')
```

```python
a = {'Movie_name':title,'Year':year,'Certificate':certificate,
                         'Runtime':duration,'Genre':genre,'Rating':rating,'Metascore':metascore,'Director':director,
                         'Votes':vote,'Gross_sale':gross}
df = pd.DataFrame.from_dict(a, orient='index')
df = df.transpose()
```
